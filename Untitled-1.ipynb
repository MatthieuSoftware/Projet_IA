Génération d'un bruit audio sinusoïdal avec du bruit. Puis affiche d'une partie de son sous la forme de graphique oscilant.

```python
#importation de fichier 

import numpy as np 
import matplotlib.pyplot as plt 
from scipy.io.wavfile import write

# Fichier de sortie où l'audio sera sauvegardé 
output_file = 'generated_audio.wav'

 # Spécifier les paramètres audio 
duration = 4  # en secondes 
sampling_freq = 44100  # en Hz 
tone_freq = 784  # en Hz (note C5) 
min_val = -4 * np.pi 
max_val = 4 * np.pi

 # Générer le signal audio 
t = np.linspace(min_val, max_val, duration * sampling_freq) 
signal = np.sin(2 * np.pi * tone_freq * t)

 # Ajouter du bruit au signal 
noise = 0.5 * np.random.rand(duration * sampling_freq) 
signal += noise

# Mettre à l'échelle aux valeurs entières sur 16 bits  
scaling_factor = np.power(2, 15) - 1 
signal_normalized = signal / np.max(np.abs(signal)) 
signal_scaled = np.int16(signal_normalized * scaling_factor)

# Sauvegarder le signal audio dans le fichier de sortie  
write(output_file, sampling_freq, signal_scaled)

 # Extraire les 200 premières valeurs du signal audio  
signal = signal[:200]

# Construire l'axe temporel en millisecondes
time_axis = 1000 * np.arange(0, len(signal), 1) / float(sampling_freq)

plt.plot(time_axis, signal, color='purple') 
plt.xlabel('Temps (millisecondes)') 
plt.ylabel('Amplitude') 
plt.title('Signal audio généré') 
plt.show()

```
